```python
textgen = textgenrnn()
textgen.train_from_file(
    'gen_trainorig_dem.txt',
    rnn_size=          128,
    rnn_layers=        3,
    rnn_bidirectional= True,
    line_delimited=    True,
    max_length=        60,
    max_words=         10000,
    batch_size =       1024,
    num_epochs=        40,
    gen_epochs=        10,
    dim_embeddings=    32,
    word_level=        True,
    train_size=        1,
    name='TextGenDem1')
textgen.save('TextGenDem.hdf5')
```
